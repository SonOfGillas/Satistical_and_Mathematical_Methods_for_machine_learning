```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Definizione delle funzioni
def f(x):
    """La funzione: f(x) = x^3 - 3x^2 + 1"""
    return x**3 - 3*x**2 + 1

def f_prime(x):
    """La prima derivata: f'(x) = 3x^2 - 6x"""
    return 3*x**2 - 6*x

def f_double_prime(x):
    """La seconda derivata: f''(x) = 6x - 6"""
    return 6*x - 6

# 2. Generazione dei valori x e calcolo delle y
x = np.linspace(-2, 4, 400)
y = f(x)
y_prime = f_prime(x)
y_double_prime = f_double_prime(x)

# 3. Creazione del grafico
plt.figure(figsize=(10, 7))

# Plot delle funzioni
plt.plot(x, y, label=r'$f(x) = x^3 - 3x^2 + 1$', color='blue')
plt.plot(x, y_prime, label=r"$f'(x) = 3x^2 - 6x$ (Radici: Min/Max)", color='red', linestyle='--')
plt.plot(x, y_double_prime, label=r"$f''(x) = 6x - 6$ (Test Min/Max)", color='green', linestyle=':')

# --- Evidenziazione Massimo Locale a x=0 ---
x_max = 0
y_max = f(x_max)
y_double_prime_max = f_double_prime(x_max)
plt.plot(x_max, y_max, 's', color='darkviolet', markersize=8, label=r'Max Locale: $f(0)=1$')
plt.plot(x_max, y_double_prime_max, 's', color='darkviolet', markersize=8) # f''(0)
plt.annotate(r'$f''(0)=-6 < 0$ (Max)', 
             xy=(x_max + 0.1, y_double_prime_max + 1), 
             color='darkviolet')


# --- Evidenziazione Minimo Locale a x=2 ---
x_min = 2
y_min = f(x_min)
y_double_prime_min = f_double_prime(x_min)
plt.plot(x_min, y_min, 'o', color='orange', markersize=8, label=r'Min Locale: $f(2)=-3$')
plt.plot(x_min, y_double_prime_min, 'o', color='orange', markersize=8) # f''(2)
plt.annotate(r'$f''(2)=6 > 0$ (Min)', 
             xy=(x_min + 0.1, y_double_prime_min - 2), 
             color='orange')

# Controlli assi
plt.axhline(0, color='black', linewidth=0.8, alpha=0.6)
plt.axvline(0, color='black', linewidth=0.8, alpha=0.6)
plt.grid(True, linestyle='--', alpha=0.4)
plt.legend()
plt.ylim(min(y) - 5, max(y) + 5) 

# Salvataggio del grafico
plt.savefig('local_min_max_test_plot.png')
```
