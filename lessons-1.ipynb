```python
# Initialize a random array of length 100_000
l = [random.gauss() for _ in range(100_000)]

# Start timer
start_time = time.time()

# Cycle on l and project to 0 all the negative elements
for i in range(len(l)):
    if l[i] < 0:
        l[i] = 0

# End timer
end_time = time.time()

# Print execution time
print(end_time - start_time)
```

```python
# Element for element operations
# it happen for: +, -, /, *, **

v = np.array([1,0,-1])
w = np.array([-1,1,2])

# to perform dot product:
print(v@w) # siccome sono entrambi a dimensionali fa finta che il primo sia una riga ed il secondo una colonna
print(np.dot(v,w))
print(np.inner(v,w))
print(v.dot(w))
```
